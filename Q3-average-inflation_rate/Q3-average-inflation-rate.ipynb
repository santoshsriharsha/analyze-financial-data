**What was the average inflation rate when their subscriptions were renewed?**

```python
import pandas as pd
```

```python
subscription_df = pd.read_csv('../data/subscription_information.csv')
```

```python
subscription_df
```

Filtering all the clients who have renewed their subscription

```python
renewed_subscription_df = subscription_df[subscription_df['renewed'] == True]
```

```python
renewed_subscription_df
```

```python
renewed_subscription_df.shape
```

```python
financial_df = pd.read_csv('../data/financial_information.csv')
```

```python
financial_df
```

```python
financial_df = financial_df.drop(['Unnamed: 0'], axis = 1)
```

```python
financial_df
```

```python
renewed_subscription_df.reset_index(drop = True)
financial_df.reset_index(drop = True)
```

```python
Joining renewed_subscription_df with financial_df to get the inflation rate when renewal 
```

```python
merged_df = renewed_subscription_df.merge(financial_df, how = 'cross')
```

```python
merged_df
```

```python
merged_df = merged_df[
    (merged_df['start_date_x'] >= merged_df['start_date_y']) &
    (merged_df['start_date_x'] <= merged_df['end_date_y'])
]
merged_df = merged_df.reset_index(drop = True)
merged_df
```

```python
average_inflation_rate = float(round(merged_df['inflation_rate'].mean(), 2))
```

```python
from IPython.display import display, Markdown
display(Markdown(f'The **average inflation rate** is {average_inflation_rate}.'))
```
